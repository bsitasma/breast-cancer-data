```python
import pandas as pd

# Load the dataset
df = pd.read_csv("breast-cancer.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Display basic information about the dataset
df.info()

# Display statistical summary of the dataset
df.describe()

# Display the number of unique values in each column
df.nunique()
```

```python
# Check for missing values
df.isnull().sum()

# Drop duplicate rows, if any
df.drop_duplicates(inplace=True)

# Verify that there are no duplicate rows left
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of diagnosis
plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=df)
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the relationship between radius_mean and area_mean
plt.figure(figsize=(10, 6))
sns.scatterplot(x='radius_mean', y='area_mean', hue='diagnosis', data=df)
plt.title('Relationship between Radius Mean and Area Mean')
plt.xlabel('Radius Mean')
plt.ylabel('Area Mean')
plt.show()
```
